```
import pandas as pd
```

```
adult = pd.read_csv('https://raw.githubusercontent.com/arx-deidentifier/arx/master/data/adult.csv', sep =';')
adult.head()
```

```
adult.info()
```

```
adult.describe(include='all')
```

```
adult['sex'] = pd.get_dummies(adult['sex'], drop_first=True)
```

```
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
adult['race']=le.fit_transform(adult['race'])
adult['marital-status']=le.fit_transform(adult['marital-status'])
adult['education']=le.fit_transform(adult['education'])
adult['native-country']=le.fit_transform(adult['native-country'])
adult['workclass']=le.fit_transform(adult['workclass'])
adult['occupation']=le.fit_transform(adult['occupation'])
```

```
adult.head()
```

Pour la prédiction, vous devrez mettre en place un régression logistique (pour laquelle il y a tout plein de paramètres https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html).

```
from sklearn.model_selection import train_test_split
X = adult.drop('salary-class', axis = 1)
y = adult['salary-class']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 42)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```
LRclf = LogisticRegression()
LRclf.fit(X_train,y_train)
```

```
LRprediction = LRclf.predict(X_test)
```

```
LRprediction
```

```
values_solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
values_iter = [50,100,200,500]
param_grid = {'solver':values_solver,'max_iter':values_iter}
grid = GridSearchCV(estimator=LRclf, param_grid=param_grid, scoring='accuracy')
grid.fit(X_train,y_train)
```

```
print(grid.best_params_)
```

```
print(grid.best_score_)
```
