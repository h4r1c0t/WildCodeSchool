```
import pandas as pd
```

```
df_people = pd.read_csv('https://raw.githubusercontent.com/arx-deidentifier/arx/master/data/adult.csv', sep=';')
```

```
df_people.head(20)
```

```
#prepocessing

colonne_sex = pd.get_dummies(df_people['sex'])
colonne_sex
```

```
colonne_marital = pd.get_dummies(df_people['marital-status'])
colonne_marital
```

```
colonne_workclass = pd.get_dummies(df_people['workclass'])
colonne_workclass
```

```
colonne_salaryclass = pd.get_dummies(df_people['salary-class'])
colonne_salaryclass
```

```
df_new = pd.concat([df_people, colonne_salaryclass, colonne_workclass, colonne_marital, colonne_sex], axis=1)
```

```
df_new
```

```
from sklearn.linear_model import LogisticRegression

X = df_new.drop(['sex', 'salary-class',	'age',	'race',	'marital-status',	'education',	'native-country',	'workclass',	'occupation',	'salary-class',	'<=50K', '>50K'], axis=1)
y = df_new['salary-class']
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```
LR = LogisticRegression()
LR.fit(X_train, y_train)
```

```
LR.score(X_test, y_test)
```

```
from sklearn.model_selection import GridSearchCV
```

```
parameters = {"penalty":['l1', 'elasticnet', 'none'], 'C':[0, 0.1, 0.5, 0.9, 1.0]}
    
#(penalty='l2', *, dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, solver='lbfgs', max_iter=100, multi_class='auto', verbose=0, warm_start=False, n_jobs=None, l1_ratio=None)}
#parametres_grid = {'min_samples_leaf' :(1, 3, 5, 10),'max_depth': (2, 3, 5, 10)}
```

```
LR2 = GridSearchCV(LR, parameters, cv = 3, scoring = 'roc_auc', verbose = 3)
LR2.fit(X_train, y_train)
```

```
LR2.best_score_
```

```
LR2.best_params_
```

```
LR3 = LogisticRegression(C= 0, penalty= 'none')
LR3.fit(X_train, y_train)
```

```
LR3.score(X_test, y_test)
```

```
y_pred = LR3.predict(X_test)
```

```
y_pred
```
