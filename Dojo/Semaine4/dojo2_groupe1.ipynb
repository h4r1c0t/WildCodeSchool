# Dictionnaires
Générez 100 nombres aléatoire compris entre 0 et 20.  
Créez un dictionnaire qui va contenir les nombres et le leur nombre d'occurences.   
- ex : pour `[4,3,4,4,1] --> {1:1, 3:1 , 4:3}`

```
import random 

dico = {}
listeNb =[]
for i in range(100):
	value = random.randint(0,20)
	listeNb.append(value)

for i in listeNb:
	dico[i] = listeNb.count(i)
 
# print(listeNb)
print(dico)
 
 
```

- Créer 3 méthodes de ton choix pour lui permettre d'effectuer des calculs
- Fait en sorte q'un objet statisticien hérite de la classe mathématicien

```
class Mathematician: 
  # attributs
  def __init__(self, nom, prenom, age):
    print( "L'expérience est le nom que chacun donne à ses erreurs. O.Wilde")
    self.nom = nom
    self.prenom = prenom
    self.age = age

  # methods
  def square(nb): 
    nb = nb**2
    return nb

  def root_square(nb):
    nb = nb**0.5
    return nb

  def puissance(nb, pow):
    result = nb**pow 
    return result


```
