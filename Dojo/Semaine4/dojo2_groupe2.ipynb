# Dictionnaires
Générez 100 nombres aléatoires compris entre 0 et 20.  
Créez un dictionnaire qui va contenir les nombres et le leur nombre d'occurences.   
- ex : pour `[4,3,4,4,1] --> {1:1, 3:1 , 4:3}`

```
import random
liste = [ random.randint(0,20) for x in range(0,100)]
compte = dict([(k, liste.count(k)) for k in set(liste)])
print(liste)
print(compte)
```

- Donne lui 3 attributs de ton choix
- Créer son constructeur de manière à ce que quand on le lance l'objet dit : "*L'expérience est le nom que chacun donne à ses erreurs. O.Wilde*"
- Créer 3 méthodes de ton choix pour lui permettre d'effectuer des calculs
- Fait en sorte q'un objet statisticien hérite de la classe mathématicien

```
class mathématicien:
  def __init__ (self, nom, prenom, age):
    self.nom=str(nom)
    self.prenom=str(prenom)
    self.age=int(age)
    print("L'expérience est le nom que chacun donne à ses erreurs. O.Wilde")

  def addition(self):
    a=int(input())
    b=int(input())
    print(a+b)

  def soustraction(self):
    c=int(input())
    d=int(input())
    print(c-d)

  def multiplication(self):
    e=int(input())
    f=int(input())
    print(e*f)

Daphné = mathématicien("Ozbek","Daphné",27)


Daphné.soustraction()
```
