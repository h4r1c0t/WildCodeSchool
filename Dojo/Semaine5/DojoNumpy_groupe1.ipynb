```
nb_sim = 10000

result_somm = np.empty(nb_sim)

for i in range(nb_sim):
  result_somm[i]=sum(np.random.choice(notes,2, replace=False))

result_somm
sum(result_somm>20)/len(result_somm)
```

```
import matplotlib.pyplot as plt
```

```
_ = plt.hist(result_somm, bins = 20)
# moyenne
_ = plt.axvline(np.mean(result_somm), color='r')
# 5%
_ = plt.axvline(np.percentile(result_somm, 5), color='g')
# 95%
_ = plt.axvline(np.percentile(result_somm, 95), color='g')
_ = plt.xlabel("Sommes des 2 notes")
plt.show()
```
