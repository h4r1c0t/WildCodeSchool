```
nombre_lancer = 10000
total = np.empty(nombre_lancer)
for i in range(nombre_lancer):
  index = np.random.randint(25)
  note1 = notes[index]
  index2 = np.random.randint(25)
  while index == index2 :
    index2 = np.random.randint(25)
  note2 = notes[index]
  if note1 + note2 > 20:
    total[i]=1
  elif note1 + note2 < 20:
    total[i]=0
prob = sum(total)/10000
print(prob)
```

```
nombre_lancer = 100000
total = np.empty(nombre_lancer)

for i in range(nombre_lancer):
  total[i] = sum(np.random.choice(notes, 2, replace=False))

sum(total>20)/nombre_lancer
```

```
import matplotlib.pyplot as plt

_ = plt.hist(total, bins = 20)
# moyenne
_ = plt.axvline(np.mean(total), color='r')
# 5%
_ = plt.axvline(np.percentile(total, 5), color='g')
# 95%
_ = plt.axvline(np.percentile(total, 95), color='g')
_ = plt.xlabel("Sommes des 2 notes")
plt.show()
```
