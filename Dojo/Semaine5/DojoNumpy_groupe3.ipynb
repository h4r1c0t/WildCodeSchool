```
result = np.random.randint(0,21,25)
print(result)
```

```
nb_simulations = 10000
sommes = np.empty(nb_simulations)

for i in range(nb_simulations):
  sommes[i] = np.sum(np.random.choice(result, 2, replace = False))

np.sum(sommes > 20)/nb_simulations
```
