```
import numpy as np

mes_notes = np.random.randint(0,21, 25)
```

```
nb_simulations = 10000
mes_sommes = np.empty(nb_simulations)

for i in range(nb_simulations):
  mes_sommes[i] = np.sum(np.random.choice(mes_notes, 2, replace=False))

print(mes_sommes)
```

```
np.sum(mes_sommes > 20)/len(mes_sommes)
```

```
import matplotlib.pyplot as plt

_ = plt.hist(mes_sommes)
_ = plt.xlabel("Somme de 2 notes")
_ = plt.axvline(np.mean(mes_sommes), color = 'r')
_ = plt.axvline(np.percentile(mes_sommes, 5), color = 'g')
_ = plt.axvline(np.percentile(mes_sommes, 95), color = 'g')
plt.show()
```
