### Vous allez travailler sur un jeu de données immobilières de San Fransisco et New York.   
### La variable in_sf vous permet de savoir si le logement est situé à San Francisco *(1= San Francisco, 0= New York)*.
---

```
url = "https://raw.githubusercontent.com/jadeyee/r2d3-part-1-data/master/part_1_data.csv"
```

##### Dans un premier temps faites les vérifications d'usage (valeurs manquantes, taille de la table, valeurs min/max, types des données).

```
import pandas as pd

data = pd.read_csv(url, sep=",", skiprows=[0, 1])
```

```
data.describe()
```

```
data.info()
```

```
data.head(10)
```

##### Puis vous réaliserez une analyse graphique du prix et de la surface des logements pour les 2 villes confondues (avec matplotlib uniquement).

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10,10))
_ = plt.scatter(data['sqft'], data['price'])
plt.title('Prix par surface')
```

##### Enfin vous ferez ressortir les différences entre les 2 villes grâce à 2 visualisations avec matplotlib uniquement (vous pouvez utiliser toutes les variables qui sont fournies dans la table).

```
import matplotlib.patches as mpatches

colors = {0: 'b', 1: 'red'}
fig = plt.figure(figsize = (10,10))
_ = plt.scatter(data['sqft'], data['price'], c=data["in_sf"].apply(lambda x: colors[x])) 
plt.title('Prix par surface')
plt.xlabel('Surface')
plt.ylabel('Prix')
red_patch = mpatches.Patch(color='red', label='NY')
green_patch = mpatches.Patch(color='b', label='SF')
plt.legend(handles=[red_patch, green_patch])
```

```
plt.figure(figsize=(10, 10))
plt.hist(data["price"])
```

```
fig = plt.figure(figsize=(10, 10))
plt.hist("sqft", data = data[data['in_sf'] == 0], color='red')
plt.hist("sqft", data = data[data['in_sf'] == 1 ], color= 'b', alpha = 0.5)
red_patch = mpatches.Patch(color='red', label='NY')
blue_patch = mpatches.Patch(color='blue', label='SF')
plt.legend(handles=[red_patch, blue_patch])
plt.title('Répartition surface par ville')
plt.xlabel('Surface')
plt.ylabel("Nombre d'occurence")
fig.tight_layout()
```

```
fig = plt.figure(figsize=(10, 10))
plt.hist("price", data = data[data['in_sf'] == 0], color='red')
plt.hist("price", data = data[data['in_sf'] == 1 ], color= 'b', alpha = 0.5)
red_patch = mpatches.Patch(color='red', label='NY')
blue_patch = mpatches.Patch(color='blue', label='SF')
plt.legend(handles=[red_patch, blue_patch])
fig.tight_layout()
```
