```
fig, ax = plt.subplots(nrows =1,ncols=2, figsize = (20,8))

ax1 = plt.subplot(121)
ax1.hist(datas['price'], bins = 492)
ax1.set_title("Prix")


ax2 = plt.subplot(122)
ax2.hist(datas['sqft'], bins = 492)
ax2.set_title("Surface")


fig.show()
```

```
plt.figure(figsize = (10, 8))

```

```
fig, ax = plt.subplots(figsize = (10, 5))

plt.scatter(datas['sqft'], datas['price'])
plt.title('Représentation du prix en fonction de la surface')

plt.show()
```

##### Enfin vous ferez ressortir les différences entre les 2 villes grâce à 2 visualisations avec matplotlib uniquement (vous pouvez utiliser toutes les variables qui sont fournies dans la table).

```
datas.head(2)
```

```
fig, ax = plt.subplots(figsize = (10, 5))

```

```
NY = datas[datas['in_sf']==0]
NY
```

```
SF = datas[datas['in_sf']==1]
SF
```

```
fig, ax = plt.subplots(figsize = (20,8))

ax1 = plt.subplot(111)
ax1.scatter(NY['sqft'], NY['price'])
ax1.set_title("Prix")
ax1.scatter(SF['sqft'], SF['price'], c='red')
ax1.legend()
plt.show()
```

```
fig, ax = plt.subplots(figsize = (20,8))

ax1 = plt.subplot(111)
ax1.scatter(NY['elevation'], NY['price_per_sqft'], label='New York')
ax1.set_title("Prix au metre carré selon la hauteur de construction")
ax1.scatter(SF['elevation'], SF['price_per_sqft'], c='red', label='San Francisco')
ax1.legend()
plt.show()
```
