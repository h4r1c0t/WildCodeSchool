##### Puis vous réaliserez une analyse graphique du prix et de la surface des logements pour les 2 villes confondues (avec matplotlib uniquement).

##### Enfin vous ferez ressortir les différences entre les 2 villes grâce à 2 visualisations avec matplotlib uniquement (vous pouvez utiliser toutes les variables qui sont fournies dans la table).

```
df0=df[df['in_sf']==0]
df0
```

```
df1=df[df['in_sf']==1]
```

```
plt.scatter(df0['sqft'], df0['price'],c='red')
plt.scatter(df1['sqft'], df1['price'],c='blue')
plt.show
```

```
plt.scatter(df0['price_per_sqft'], df0['elevation'],c='red')
plt.scatter(df1['price_per_sqft'], df1['elevation'],c='blue')
plt.legend(['New-York','San Fransisco'])
plt.title('Comparaison prix au m2 et étage')
plt.xlabel('m2')
plt.ylabel('m2')
plt.show()
```
