##### Puis vous réaliserez une analyse graphique du prix et de la surface des logements pour les 2 villes confondues (avec matplotlib uniquement).

```
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.scatter(df.index,df["price"])
plt.ylabel("prix")
plt.xlabel('logements')
plt.title('Prix des logements')
plt.show()

plt.figure(figsize=(20,5))
plt.scatter(df.index, df["sqft"])
plt.ylabel("surface")
plt.xlabel('logements')
plt.title('Surface des logements')
plt.show()
```
