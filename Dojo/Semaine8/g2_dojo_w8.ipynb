```
import pandas as pd
import numpy as np
link = "https://raw.githubusercontent.com/briatte/aad/master/reinhart-rogoff/debt.csv"
df = pd.read_csv(link,index_col=0)
```

Faire une représentation graphique de la croissance en fonction du ratio dette/PIB.

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.bar(df['ratio'], df['growth'])

plt.show()
```

Choisir 5 pays, faire un scattter plot qui représente la croissance en fonction du ratio et colorer les points par pays.

```
df['Country'].unique()
```

```
df5=df[(df['Country']=='Australia')|(df['Country']=='Belgium')|(df['Country']=='Canada')|(df['Country']=='Danemark')|(df['Country']=='Finland')]
df5
```

```
plt.figure(figsize = (10,8))
plt.title('Croissance par pays en fonction du ratio')
sns.scatterplot(x = 'ratio', y = 'growth', data = df5, hue = 'Country')
```

Réaliser un box plot de la croissance pour ces 5 pays.

```
plt.figure(figsize = (10,8))
sns.boxplot(x='growth', data = df5, y= 'Country')
```

A l'aide numpy réaliser une régression linéaire avec la croissance en y (variable expliquée) et le ratio en x (variable explicative).   
Représenter la droite de régression sur le graphique.

```
import numpy as np
```

```
# You have to define your Min and Max value for the X-axis
Xmin = df['ratio'].min()
Xmax = df['ratio'].max()            # Complete this code

# We generate an array with a range of values between your min and max values
array = np.arange(Xmin, Xmax)
model = np.polyfit(x = df['ratio'], y = df['growth'], deg = 1)


plt.scatter(x=df['ratio'], y=df['growth'], color='blue')
plt.plot(array, np.poly1d(model)(array), color='red')
plt.show()
```
