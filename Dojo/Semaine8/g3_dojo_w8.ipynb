Importer les données situées à l'adresse suivante:   
https://raw.githubusercontent.com/briatte/aad/master/reinhart-rogoff/debt.csv

```
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/briatte/aad/master/reinhart-rogoff/debt.csv')
df.head()
```

Faire les vérifications d'usage (taille des données, nombre de valeurs manquantes, type des variables)

```
df.info()
```

Faire une représentation graphique de la croissance en fonction du ratio dette/PIB.

```
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df['growth'],df['ratio'])
plt.title("Croissance en fonction du ratio dette/PIB")
plt.xlabel('Croissance')
plt.ylabel('Ratio Dette/PIB')
```

Choisir 5 pays, faire un scattter plot qui représente la croissance en fonction du ratio et colorer les points par pays.

```
countries = ['Australia', 'Belgium', 'Denmark', 'France', 'Greece']
df2 = df[df['Country'].isin(countries)]
import seaborn as sns

_ = sns.scatterplot(data = df2, x = df2['growth'], y =df2['ratio'],  hue = 'Country')
```

Réaliser un box plot de la croissance pour ces 5 pays.

```
_ = sns.boxplot(x = 'Country', y ='growth', data=df2)
```

A l'aide numpy réaliser une régression linéaire avec la croissance en y (variable expliquée) et le ratio en x (variable explicative).   
Représenter la droite de régression sur le graphique.

```
import numpy as np

Xaxis = df['ratio']
Yaxis = df['growth']
model = np.polyfit(x = Xaxis, y = Yaxis, deg = 1)
Xmin = Xaxis.min()
Xmax = Xaxis.max()
array = np.arange(Xmin, Xmax)
plt.scatter(Xaxis,Yaxis,c='b')
plt.plot(Xaxis, np.poly1d(model)(Xaxis),c='r')

plt.show()
```
