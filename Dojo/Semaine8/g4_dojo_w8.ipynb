```
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/briatte/aad/master/reinhart-rogoff/debt.csv")
```

Faire une représentation graphique de la croissance en fonction du ratio dette/PIB.

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
plt.figure(figsize=(10,10))
plt.scatter(data['growth'], data['ratio'])
plt.show()
```

```
plt.figure(figsize=(20,10))
sns.lineplot(y='growth', x='ratio', data=data)
plt.show()
```

Choisir 5 pays, faire un scattter plot qui représente la croissance en fonction du ratio et colorer les points par pays.

```
data["Country"].unique()
df_countries = data[data["Country"].isin(["France","Canada","Italy","Australia","Belgium"])]
df_countries
sns.scatterplot( x='ratio', y='growth', hue='Country', data=df_countries)
```

Réaliser un box plot de la croissance pour ces 5 pays.

```
sns.boxplot(x='Country', y='growth', data=df_countries)
```

A l'aide numpy réaliser une régression linéaire avec la croissance en y (variable expliquée) et le ratio en x (variable explicative).   
Représenter la droite de régression sur le graphique.

```


plt.figure(figsize=(20,10))
sns.regplot(x='ratio', y='growth', data=df_countries)
```

```
import numpy as np

Xaxis = df_countries['ratio']
Yaxis = df_countries['growth']
model = np.polyfit(x = Xaxis, y = Yaxis, deg = 1)
Xmin = Xaxis.min()
Xmax = Xaxis.max()
array = np.arange(Xmin, Xmax)

fig, ax = plt.subplots(figsize = (10, 4))
_ = plt.scatter('ratio', 'growth', data = df_countries, c = '#9BC53D')
_ = plt.plot(array, np.poly1d(np.polyfit(x = Xaxis, y = Yaxis, deg = 1))(array), c = '#C0392B', alpha = .5)
plt.show()
```
