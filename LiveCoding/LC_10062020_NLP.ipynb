```
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
exemple = tokenizer.tokenize(exemple)
```

```
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')


stop_words = set(stopwords.words('english')) 
filtered_exemple = [w for w in exemple if not w in stop_words] 
```

```
from nltk.stem import PorterStemmer

ps = PorterStemmer() 
filtered_exemple = [ps.stem(w) for w in filtered_exemple]
```

```
#filtered_exemple
```

```
from nltk.stem import WordNetLemmatizer 

nltk.download('wordnet')
lemmatizer = WordNetLemmatizer() 
filtered_exemple = [lemmatizer.lemmatize(w) for w in filtered_exemple]
#filtered_exemple
```

```
X = ' '.join(filtered_exemple)
```

```
corpus = []
corpus.append(X)
```

```
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
```

```
corpus = []

for i in range(df_reviews.shape[0]-45000):
  soup = BeautifulSoup(df_reviews.iloc[i][0], "html.parser")
  review = soup.get_text()
  review = review.lower()
  review = tokenizer.tokenize(review)
  review = [w for w in review if not w in stop_words] 
  review = [ps.stem(w) for w in review]
  review = [lemmatizer.lemmatize(w) for w in review]
  review = ' '.join(review)
  corpus.append(review)
```

```
pd.DataFrame(X.toarray(), columns = vectorizer.get_feature_names())

#print(vectorizer.get_feature_names())
```

```
y = df_reviews['sentiment'][:5000]
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```
from sklearn.svm import LinearSVC

linear_svc = LinearSVC(C=0.5, random_state=42)
linear_svc.fit(X_train, y_train)

y_pred = linear_svc.predict(X_test)
```

```
y_pred
```

```
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```
