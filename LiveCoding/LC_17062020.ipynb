```python
names = df_metadata['name'].unique()
names
```

```python
for name in names:
    query = "select * from " + name
    tb = pd.read_sql(query, conn)
    print(name, tb.columns,tb.shape, '\n')
```

```python
dict_metadata = {'nom_table':[],'noms_colonnes':[],'nombre_lignes':[], 'nombres_colonnes':[]}

for name in names:
    query = "select * from " + name
    tb = pd.read_sql(query, conn)
    dict_metadata['nom_table'].append(name)
    dict_metadata['noms_colonnes'].append(tb.columns)
    dict_metadata['nombre_lignes'].append(tb.shape[0])
    dict_metadata['nombres_colonnes'].append(tb.shape[1])
```
