# Python et algorithmie

Ecris une fonction qui prend un nombre entier compris entre 1 et 300 inclus, et le convertit en chiffre romains.

https://jeretiens.net/wp-content/uploads/2015/12/les_chiffres_romains.jpg
![Exemple de chiffres romains](https://jeretiens.net/wp-content/uploads/2015/12/les_chiffres_romains.jpg)

````
# Ta fonction ici :
def intToRom(nb):
  import numpy as np
  # décomposition du nombre en centaines, dizaines et unités.
  centaine = nb //100
  dizaine = nb // 10 - centaine*10
  unite = nb - dizaine*10 - centaine*100
  # table de correspondance des nombre romains 
  roms = np.array([['', 'I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX'], 
                   ['', 'X', 'XX', 'XXX', 'XL', 'L', 'LX', 'LXX', 'LXXX', 'XC'], 
                   ['', 'C', 'CC','CCC', 'CD', 'D', 'DC', 'DCC', 'DCCC', 'CM']])
  # conversion en nombre romain
  nb_roms = roms[2][centaine] + roms[1][dizaine] + roms[0][unite]
  # affiche la correspondance entre le nombre choisi et sa conversion 
  print(nb, ':', nb_roms)
  # retourne le nombre romain
  return nb_roms

intToRom(int(input()))
````

Tu devras ajouter une méthode .clean() qui nettoyera les espaces en début et en fin de chaine, mettra en majuscule le premier caractère, et tous les autres caractères en minuscule, et ajoutera un point "." à la fin.

Tu montreras le résultats en appliquant la méthode à la phrase ci-dessous.




````
# Tu mettras création de ta classe ici :
class sentence(str):
  def __init__(self, variable):
    str.__init__(self)
    self.variable = variable

  # method
  def clean(self):
    self.variable = self.variable.strip()
    self.variable = self.variable.lower()
    self.variable = self.variable[0].upper() + self.variable[1:]
    self.variable = self.variable + '.'
    print(self.variable)

# Ce code permettra de tester :
variable = " bonjour, AUJourd'hui c'EST le cheCKpoint   "
test = sentence(variable)
test.clean()
````

#SQL

A partir de la base de données client que [tu trouveras ici](https://www.w3schools.com/sql/trysql.asp?filename=trysql_op_in), crée une requête SQL qui fournit les **noms de chaque catégorie**, le **chiffre d'affaire associé à cette catégorie**, et le **prix moyen pondéré des produits vendus pour cette catégorie**. 

Le chiffre d'affaire sera arrondi à l'unité la plus proche, et le prix moyen comportera 2 décimales.

Ce tableau devra être trié pour avoir les catégories avec le chiffre d'affaire le plus élevé en premier.

Le code SQL devra être copié / collé dans un bloc de code ci-dessous.



*** Exemple des premières lignes attendues ***
*********************

Categories | Chiffre_d_affaire | Prix_moyen
- |:-: | -:
Beverages | 99465 | 43.45
Dairy Products | 69921 | 26.88
... | ... | ...

```sql 
-- Ton code SQL ici :
SELECT cat.CategoryName AS Categories, ROUND(SUM(p.price * od.Quantity)) AS Chiffre_d_affaire, ROUND(SUM(p.price * od.Quantity)/SUM(od.Quantity), 2) AS Prix_moyen 
FROM Products AS p
JOIN OrderDetails AS od ON od.ProductID = p.ProductID
JOIN Categories AS cat ON cat.CategoryID = p.CategoryID
GROUP BY Categories
ORDER BY Chiffre_d_affaire DESC
```

````
-- Ton code SQL ici :
SELECT cat.CategoryName AS Categories, ROUND(SUM(p.price * od.Quantity)) AS Chiffre_d_affaire, ROUND(SUM(p.price * od.Quantity)/SUM(od.Quantity), 2) AS Prix_moyen 
FROM Products AS p
JOIN OrderDetails AS od ON od.ProductID = p.ProductID
JOIN Categories AS cat ON cat.CategoryID = p.CategoryID
GROUP BY Categories
ORDER BY Chiffre_d_affaire DESC
````
