# Cross Validation

#### Import modules

```
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import make_scorer, accuracy_score

from sklearn.model_selection import cross_val_score, StratifiedKFold
```

## Import dataset

```
url = "https://raw.githubusercontent.com/h4r1c0t/WildCodeSchool/master/Odyssey/Dataset/titanic.csv"
df_titanic = pd.read_csv(url)
df_titanic = pd.get_dummies(df_titanic, columns = ['Sex'], drop_first = True)
```

```
df_titanic.describe(include = 'all').T
```

```
df_titanic.head()
```

## KNN Classifier -- simple split

### Train test split

```
X = df_titanic[['Pclass', 'Sex_male', 'Age']]
y = df_titanic['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2, train_size = 0.75)
```

### Train model

```
kNNRegModel = KNeighborsClassifier().fit(X_train, y_train)

y_pred = kNNRegModel.predict(X_test)

print("Score for the Test dataset :", round(accuracy_score(y_pred, y_test), 5))
```

## KNN Classifer -- Cross Validation

```
cv = StratifiedKFold(n_splits = 6, random_state = 42, shuffle = True)

scores = cross_val_score(KNeighborsClassifier(), X, y, cv = cv, scoring = make_scorer(accuracy_score))
scores
```
