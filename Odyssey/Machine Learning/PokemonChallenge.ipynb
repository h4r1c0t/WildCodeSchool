# Challenge KNN classification : Pokemon

#### Import modules

```
import numpy as pd
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
```

#### Import dataset

```
df_pokemon = pd.read_csv('https://raw.githubusercontent.com/h4r1c0t/WildCodeSchool/master/Odyssey/Dataset/pokemon.csv')
```

```
df_pokemon.describe(include = 'all')
```

### kNN model

#### Boucle sur la liste de pokémon

```
df_pokemon_noleg = df_pokemon[df_pokemon['Legendary'] == False]
```

```
cols_predict = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
```

```
X = df_pokemon_noleg[cols_predict]
y = df_pokemon_noleg['Legendary']
```

```
pkm_kNNcls = KNeighborsClassifier(n_neighbors = 1).fit(X, y)
```

```
pkm_to_remplace = ['Mewtwo', 'Lugia', 'Rayquaza', 'Giratina Origin Forme', 'Dialga', 'Palkia']
```

```
for pkm in pkm_to_remplace:
  prdct_pkm_id = pkm_kNNcls.kneighbors(df_pokemon[df_pokemon['Name'] == pkm][cols_predict], 1)[1][0][0]
  prdct_pkm_dist = pkm_kNNcls.kneighbors(df_pokemon[df_pokemon['Name'] == pkm][cols_predict], 1)[0][0][0]
  prdct_pkm_name = df_pokemon[df_pokemon.index == prdct_pkm_id]['Name']
  print('{} could be {} with a distance of {}'.format(pkm, prdct_pkm_name.values[0], prdct_pkm_dist))
```

#### Adding Type to the model

```
df_pokemon_noleg_dummies = pd.get_dummies(df_pokemon_noleg, prefix = ['Type1', 'Type2'], columns = ['Type 1', 'Type 2'])
```

```
df_pokemon_dummies = pd.get_dummies(df_pokemon, prefix = ['Type1', 'Type2'], columns = ['Type 1', 'Type 2'])
```

```
df_pokemon_noleg_dummies.columns
```

```
cols_predict = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Type1_Bug', 'Type1_Dark', 'Type1_Dragon',
                'Type1_Electric', 'Type1_Fairy', 'Type1_Fighting', 'Type1_Fire', 'Type1_Flying', 'Type1_Ghost', 
                'Type1_Grass', 'Type1_Ground', 'Type1_Ice', 'Type1_Normal', 'Type1_Poison', 'Type1_Psychic',
                'Type1_Rock', 'Type1_Steel', 'Type1_Water', 'Type2_Bug', 'Type2_Dark', 'Type2_Dragon', 'Type2_Electric', 
                'Type2_Fairy', 'Type2_Fighting', 'Type2_Fire', 'Type2_Flying', 'Type2_Ghost', 'Type2_Grass', 'Type2_Ground', 
                'Type2_Ice', 'Type2_Normal', 'Type2_Poison', 'Type2_Psychic', 'Type2_Rock', 'Type2_Steel', 'Type2_Water']
```

```
X2 = df_pokemon_noleg_dummies[cols_predict]
y2 = df_pokemon_noleg_dummies['Legendary']
```

```
pkm_kNNcls2 = KNeighborsClassifier(n_neighbors = 1).fit(X2, y2)
```

```
for pkm in pkm_to_remplace:
  prdct_pkm_id = pkm_kNNcls2.kneighbors(df_pokemon_dummies[df_pokemon_dummies['Name'] == pkm][cols_predict], 1)[1][0][0]
  prdct_pkm_dist = pkm_kNNcls2.kneighbors(df_pokemon_dummies[df_pokemon_dummies['Name'] == pkm][cols_predict], 1)[0][0][0]
  prdct_pkm_name = df_pokemon_dummies[df_pokemon_dummies.index == prdct_pkm_id]['Name']
  print('{} could be {} with a distance of {}'.format(pkm, prdct_pkm_name.values[0], prdct_pkm_dist))
```
