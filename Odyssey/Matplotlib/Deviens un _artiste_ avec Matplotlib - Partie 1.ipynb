```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

dt = 0.01
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t))               
nse2 = np.random.randn(len(t))               

s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1)
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('time')
axs[0].yaxis.set_label_coords(1.05, 0.5)
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)
axs[0].cohere(s1, s2, 256, 1. / dt)

axs[1].plot(t, s2)
axs[1].set_xlim(0, 2)
axs[1].set_xlabel('time')
axs[1].set_ylabel('s2')
axs[1].yaxis.set_label_coords(1.05, 0.5)
axs[1].grid(True)

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt   # on importe matplotlib
import numpy as np                # on importe numpy
x = np.linspace(0, 10)            # on créé un array contenant les coordonnées sur l'axe horizontal
with plt.style.context('Solarize_Light2'):            # on applique une palette de couleur pour les éléments suivants
    plt.plot(x, np.sin(x) + x + np.random.randn(50))      # on trace la 1ère courbe avec x et y = x + un nombre aléatoire
    plt.plot(x, np.sin(x) + 2 * x + np.random.randn(50))  # on trace la 2d courbe avec x et y = 2 * x + un nombre aléatoire
    plt.plot(x, np.sin(x) + 3 * x + np.random.randn(50))  # on trace la 3ème courbe avec x et y = 3 * x + un nombre aléatoire
    plt.plot(x, np.sin(x) + 4 + np.random.randn(50))      # on trace la 4ème courbe avec x et y = 4 * x + un nombre aléatoire
    plt.plot(x, np.sin(x) + 5 * x + np.random.randn(50))  # on trace la 5ème courbe avec x et y = 5 * x + un nombre aléatoire
    plt.plot(x, np.sin(x) + 6 * x + np.random.randn(50))  # on trace la 6ème courbe avec x et y = 6 * x + un nombre aléatoire
    plt.plot(x, np.sin(x) + 7 * x + np.random.randn(50))  # on trace la 7ème courbe avec x et y = 7 * x + un nombre aléatoire
    plt.plot(x, np.sin(x) + 8 * x + np.random.randn(50))  # on trace la 8ème courbe avec x et y = 8 * x + un nombre aléatoire
    plt.title('8 Random Lines - Line') # on donne un titre au graphique
    plt.xlabel('x label', fontsize=14) # on donne un titre à l'axe des absices et on definit la taille de police
    plt.ylabel('y label', fontsize=14) # on donne un titre à l'axe des ordonnées et on definit la taille de police

plt.show()
```
