```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

people = ('Bernard', 'Hector', 'Jessica', 'Jacques', 'Martine')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center',
        color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis() 
ax.set_xlabel('Avancement')
ax.set_title("Quel est ton taux d'avancement sur les quêtes ?")

plt.show()
```

# Mission 2

Modifie le code ci-dessous dans une nouvelle cellule pour que :
- Les valeurs pour hommes soient à droite et celles pour femmes à gauche
- Modifier la couleur des colonnes
- Diminuer l'épaisseur des paires de baton

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels)) 
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

```python
import matplotlib                                           #01. importe matplotlib
import numpy as np                                          #02. importe numpy avec un alias
import matplotlib.pyplot as plt                             #03. réimporte matplotlib avec un alias (utilité ??)

np.random.seed(19680801)                                    #05. génère les "aléatoires" à partir de la graine donnée

mu = 100                                                    #07. assigne une valeur à µ
sigma = 15                                                  #08. assigne une valeur à sigma
x = mu + sigma * np.random.randn(437)                       #09. cacule des coordonnées en x dans une array avec de l'aléatoire

num_bins = 50                                               #11. définie un nombre de bins à afficher

fig, ax = plt.subplots()                                    #13. defini plans des graphiques

n, bins, patches = ax.hist(x, num_bins, density=1)          #15. trace l'histogramme de x par rapport à bins

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))           #18. définie les coordonnées de y 
ax.plot(bins, y, '--')                                      #19. trace la courbe en pointillés 
ax.set_xlabel('Smarts')                                     #20. titre de l'axe x
ax.set_ylabel('Probability density')                        #21. titre de l'axe y 
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')    #22. titre du graphique

fig.tight_layout()                                          #24. ajuste les paramètre des axes pour le rendu final
plt.show()                                                  #25. affiche le graphique
```
