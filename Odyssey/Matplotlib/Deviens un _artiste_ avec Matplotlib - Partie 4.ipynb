# Mission 1

Crée un scatter plot entre deux variables x et y de ton choix. Ajoute une légende et met les valeurs dans une couleur autre que noir.

```
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

fig, ax = plt.subplots()
ax.scatter(x, y, c='r')
ax.legend(labels = ('points',))
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt


rng = np.random.RandomState(0)
x = rng.randn(100)              # génère un array numpy avec 100 nombres entre -3 et 3 qui seront les coordonnées en absice 
y = rng.randn(100)              # génère un array numpy avec 100 nombres entre -3 et 3 qui seront les coordonnées en ordonnée
colors = rng.rand(100)          # génère un array numpy avec 100 nombres entre -3 et 3 qui représenteront les couleurs à donner aux points
sizes = 1000 * rng.rand(100)    # génère un array numpy avec 100 nombres entre 0 et 1000 qui représenteront la taille à donner aux points

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
    """
    Dans le code ci-dessus:
      - x -> coordonnées en absice des points
      - y -> coordonnées en ordonnée des points
      - c -> couleur des points
      - s -> taille des points
      - alpha -> transparence des points
      - cmap -> palette de couleur 
    """
plt.colorbar();   # trace la barre de légende avec le dégradé de couleurs
```
