```
# Import dataset
import pandas as pd

df = pd.read_csv('http://bit.ly/drinksbycountry')
```

```
df.head()
```

```
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
all_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
labels = ['x1', 'x2', 'x3']

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# rectangular box plot
bplot1 = axes[0].boxplot(all_data,
                         vert=True, 
                         patch_artist=True,  
                         labels=labels) 
axes[0].set_title('Rectangular box plot')

bplot2 = axes[1].boxplot(all_data,
                         notch=True, 
                         vert=True, 
                         patch_artist=True, 
                         labels=labels) 
axes[1].set_title('Notched box plot')

colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xlabel('Three separate samples')
    ax.set_ylabel('Observed values')

plt.show()
```

Commente le code ci-dessous pour l'expliquer à un autre Data analyst. Ensuite applique le à ce dataset : http://bit.ly/drinksbycountry

```
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4)) # Génère les paramètre de la figure (lignes, colonnes, taille)

np.random.seed(19680801)  # fixe une graine particulière pour retrouver les mêmes données "aléatoires" à chaque fois


all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]  # Génère les datas

axes[0].violinplot(all_data,            # trace un violin plot sur le premier axe
                   showmeans=False,     # n'affiche pas la moyenne
                   showmedians=True)    # mais plutôt la médiane
axes[0].set_title('Violin plot')

axes[1].boxplot(all_data)               # trace un boxplot sur le second axe
axes[1].set_title('Box plot')

for ax in axes:                         # def les paramètres pour tout les axes
    ax.yaxis.grid(True)                                   # affiche la grille
    ax.set_xticks([y + 1 for y in range(len(all_data))])  # choisit les ticks pour l'axe x
    ax.set_xlabel('Four separate samples')                # defini le titre des axes x 
    ax.set_ylabel('Observed values')                      # defini le titre des axes y 

plt.setp(axes, xticks=[y + 1 for y in range(len(all_data))],  # set les paramètres sur la figure
         xticklabels=['x1', 'x2', 'x3', 'x4'])                # nomme les ticks de x en fonction des catégories
plt.show()  # affiche les graphs
```
