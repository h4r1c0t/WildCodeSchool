# Mission 3

Recreate the visual below using Seaborn

```
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels)) 
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```

```
import pandas as pd
import seaborn as sns

data = {'labels':['G1', 'G2', 'G3', 'G4', 'G5', 'G1', 'G2', 'G3', 'G4', 'G5'], 
        'means':[20, 34, 30, 35, 27, 25, 32, 34, 20, 25],
        'sex':['men', 'men', 'men', 'men', 'men', 'women', 'women', 'women', 'women', 'women']}
df = pd.DataFrame.from_dict(data)

ax = sns.barplot('labels', 'means', hue='sex', data=df, palette = ['#006599', '#FF5D00'])

plt.show()
```

```
import pandas as pd
import seaborn as sns
sns.set()

df = sns.load_dataset("brain_networks", header=[0, 1, 2], index_col=0)  # charge un dataframe

used_networks = [1, 5, 6, 7, 8, 12, 13, 17]             
used_columns = (df.columns.get_level_values("network")    # C'est une variable avec les colonnes Ã  utiliser
                          .astype(int)                    # format int
                          .isin(used_networks))           # qui font parti de la liste used network
df = df.loc[:, used_columns]                            # Ne garde dans df que les colonnes d'intÃ©rÃªt

network_pal = sns.husl_palette(8, s=.45)      # Configure la palette pour la coloration
network_lut = dict(zip(map(str, used_networks), network_pal))   # gÃ©nÃ¨re des groupes 'network' et leut assigne une couleur de la pelette 

networks = df.columns.get_level_values("network")   # rÃ©cupÃ¨re les index des colonnes 
network_colors = pd.Series(networks, index=df.columns).map(network_lut) # gÃ©nÃ¨re un mapping des couleurs pour les colonnes

sns.clustermap(df.corr(), center=0, cmap="vlag",      # dessine la matrice des corrÃ©lations entre les variables. 
               row_colors=network_colors, col_colors=network_colors,  # coloration des lignes et colonnes
               linewidths=.75, figsize=(13, 13))      # paramÃ¨tres de la figure et des lignes de l'aboraissance
```
