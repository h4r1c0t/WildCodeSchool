# Mission 1

Reproduce the graphic below using Seaborn

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.show()
```

```
import seaborn as sns

sns.scatterplot(x, y, hue = colors, size = sizes, sizes=(min(sizes), max(sizes)), alpha = 0.3, palette = 'viridis')
plt.show()
```

# Mission 2

Reproduce the graphic below using Seaborn

```
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(19680801)

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots()
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set(xlabel='entry a', ylabel='entry b')
plt.show()
```

```
_ = sns.scatterplot('a', 'b', hue='c', size='d', sizes = (min(sizes), max(sizes)), data=data)
_.set(xlabel='entry a', ylabel='entry b')
plt.show()
```

```
value1 = [82,76,24,40,67,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52]
value2=[62,5,91,25,36,32,96,95,3,90,95,32,27,55,100,15,71,11,37,21]
value3=[23,89,12,78,72,89,25,69,68,86,19,49,15,16,16,75,65,31,25,52]
value4=[59,73,70,16,81,61,88,98,10,87,29,72,16,23,72,88,78,99,75,30]
 
box_plot_data=[value1,value2,value3,value4]
box=plt.boxplot(box_plot_data,vert=0,patch_artist=True,labels=['course1','course2','course3','course4'],
            )
 
colors = ['cyan', 'lightblue', 'lightgreen', 'tan']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
 
plt.show()
```

```
df1 = pd.DataFrame(value1)
df2 = pd.DataFrame(value2)
df3 = pd.DataFrame(value3)
df4 = pd.DataFrame(value4)

df1['course'] = 'course1'
df2['course'] = 'course2'
df3['course'] = 'course3'
df4['course'] = 'course4'

df = pd.concat([df1, df2, df3, df4])
```

```
sns.boxplot(x = 0, y = 'course', data = df, order = ['course4', 'course3', 'course2', 'course1'])
plt.show()
```

# Mission 4

Reproduce the graphic below using Seaborn

```
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

np.random.seed(19680801)


all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

axes[0].violinplot(all_data,
                   showmeans=False,
                   showmedians=True)
axes[0].set_title('Violin plot')

axes[1].boxplot(all_data)
axes[1].set_title('Box plot')

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y + 1 for y in range(len(all_data))])
    ax.set_xlabel('Four separate samples')
    ax.set_ylabel('Observed values')

plt.setp(axes, xticks=[y + 1 for y in range(len(all_data))],
         xticklabels=['x1', 'x2', 'x3', 'x4'])
plt.show()
```

```
import pandas as pd
np.random.seed(19680801)
all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

df_box = pd.DataFrame({'x':[]})

for i in range(4):
  df = pd.DataFrame(all_data[i])
  df['x'] = str('x{}'.format(i+1))
  df_box = pd.concat([df_box, df])
```

```
import seaborn as sns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

ax1 = plt.subplot(121)
ax1 = sns.violinplot(x = 'x', y = 0, data = df_box)
ax1.yaxis.set_label_text('Observed Values')
ax1.xaxis.set_label_text('Four separate samples')
ax1.set_title('Violin plot')

ax2 = plt.subplot(122)
ax2 = sns.boxplot(x = 'x', y = 0, data = df_box)
ax2.yaxis.set_label_text('Observed Values')
ax2.xaxis.set_label_text('Four separate samples')
ax2.set_title('Box plot')

plt.show()
```

# Mission 5

Apply a heatmap like the one shown below to columns of your choice from this Pokemon dataset: https://www.kaggle.com/abcsds/pokemon

```
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
plt.show()
```
