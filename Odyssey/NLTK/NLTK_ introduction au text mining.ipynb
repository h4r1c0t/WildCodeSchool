
#### 3. Transforme ton texte en tokens de phrases, puis en tokens de mots.



```
# Sentences tokenization
sentences = sent_tokenize(article)
sentences
```

```
# Words tokenization
words = word_tokenize(article)
words
```

#### 4. Compte la fréquence d'apparition de chaque mot et réalise un graphique qui représente les 20 mots les plus présents dans ton texte.



```
fDist1 = FreqDist(words)

fDist1 = sorted(fDist1.items(), key = lambda x : x[1])
fDist1 = fDist1[-20:]
fDist1.reverse()
```

```
[plt.bar(k, v) for k, v in fDist1]

plt.xticks(rotation = 90)
plt.show()
```

#### 5. Enlève les stop words et regarde à nouveau quels sont les mots les plus présents. Là encore, réalise un graphique.



```
# On filtre les stop words 
stop_words = set(stopwords.words("english"))
filtred_words = [w for w in words if not w in stop_words]

# Et aussi la ponctuation
punctuation = ['.',',','"','?',';',':','}','{','(',')','!','@','#',"''",'""', "``"]

filtred_words = [w for w in filtred_words if not w in punctuation]

fDist2 = FreqDist(filtred_words)

fDist2 = sorted(fDist2.items(), key = lambda x : x[1])
fDist2 = fDist2[-20:]
fDist2.reverse()
```

```
[plt.bar(k, v) for k, v in fDist2]

plt.xticks(rotation = 90)
plt.show()
```

#### 6. Utilise un stemmer et/ou un lemmatizer pour rendre le corpus de mots encore plus pertinent, et réalise un dernier graphique.



```
ps = PorterStemmer()
filtred_words2 = []
for w in filtred_words:
  word = ps.stem(w)
  if not word in stop_words:
    filtred_words2.append(word)
```

```
fDist3 = FreqDist(filtred_words2)

fDist3 = sorted(fDist3.items(), key = lambda x : x[1])
fDist3 = fDist3[-20:]
fDist3.reverse()
```

```
[plt.bar(k, v) for k, v in fDist3]

plt.xticks(rotation = 90)
plt.show()
```

```
lemmatizer = WordNetLemmatizer()
filtred_words3 = []
for w in filtred_words2:
  word = lemmatizer.lemmatize(w)
  if not word in stop_words:
    filtred_words3.append(word)
```

```
fDist4 = FreqDist(filtred_words3)

fDist4 = sorted(fDist4.items(), key = lambda x : x[1])
fDist4 = fDist4[-20:]
fDist4.reverse()
```

```
[plt.bar(k, v) for k, v in fDist4]

plt.xticks(rotation = 90)
plt.show()
```

#### 7. Sélectionne le preprocessing le plus pertinent et réalise un wordcloud mettant en avant les mots les plus utilisés dans ton texte.



```
# On garde la dernière méthode et on créé une phrase avec les mots
text = ""
for w in filtred_words3:
  text += ' '
  text += w

wordcloud = WordCloud(width=254, height=128, max_font_size=50, max_words=100, 
                      background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()        # Je ne sais pas pourquoi mais j'ai des doublons de mots... 
```
