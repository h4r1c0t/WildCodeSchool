## Mission 2

Creates a function with Numpy that indicates for which value of x, between -100 and 100, cosine and sine are equal.

```
array = np.random.randint(-100, 100, 1000)

def cosNsin(array):
  sin = []
  cos = []
  for i in range(array.size):
    sin.append(round(np.sin(array[i] * np.pi / 180.), 5)) # calcul du sinus de chaque valeur de array
    cos.append(round(np.cos(array[i] * np.pi / 180.), 5))  # calcul du cosinus de chaque valeur de array
  cosSin = np.array([cos, sin])
  equalities = {}
  index = []
  for i in range(array.size):
    if cosSin[0][i] == cosSin[1][i]:
      index.append(i)
      equalities[array[i]] = index
  return equalities

cosNsin(array)
```

## Mission 3

Creates a function that returns the [inverse value](https://en.wikipedia.org/wiki/Multiplicative_inverse) of a set of values contained in a Numpy array

```
def inverse(array):
  array = array.astype(float)
  for i in range(array.size):
    array[i] = 1/array[i]
  return array

array1 = np.random.randint(0, 100, 20)
print(array1, '\n')
print('inverse: \n', inverse(array1))
```

## Mission 4

Create a function with Numpy that solves [second degree equations of the kind](https://www.youtube.com/watch?v=hwq_zpsGixc): $ax^2+bx+c$

```
def Eq2d(a, b, c):
  return np.roots(np.array([a, b, c]))

print(Eq2d(1, 10, 21), "\n")
print(Eq2d(1, -3, -10))
```

## Mission 5

Recodes the central limit theorem using Numpy functions and not Python functions.

```
# Code with Python

import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp

domaine = range(-100,100)
mu = 0
sigma = 20

f = lambda x : 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-mu),2)/(2*pow(sigma,2)))

y = [f(x) for x in domaine]
plot = plt.plot(domaine, y)
```

```
# Code with Numpy

import numpy as np
import matplotlib.pyplot as plt

domaine = range(-100,100)
mu = 0
sigma = 20

f = lambda x : 1/(np.sqrt(2*np.pi*pow(sigma,2))) * np.exp(-pow((x-mu),2)/(2*pow(sigma,2)))

y = [f(x) for x in domaine]
plot = plt.plot(domaine, y)
```
