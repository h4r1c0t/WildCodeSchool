## Mission 13: 
- Create 2 arrays, randomly fill them with 20 int type values between 0 and 10.
- Create a function that will run through each array with a while loop.
- At each turn of the loop the function will compare the values of the 2 arrays located at the same position.
- In order to return the index to which the values of the 2 arrays are equal.
- If no value is equal to the same position in the 2 arrays, return the message: "No value in common".



```
array1 = np.random.randint(0, 10, 20)
array2 = np.random.randint(0, 10, 20)

def arrays_comparison(arr1, arr2):
  i = 0
  it = max(arr1.size, arr2.size)
  liste = []
  while i <= it -1:
    if arr1[i] == arr2[i]:
      liste.append(i)
    i += 1
  if len(liste) != 0: 
    return "Common values at index:", liste
  else:
    return "No common values"

arrays_comparison(array1, array2)
```
