# Mission 5

Renames the columns of the dataset below http://bit.ly/uforeports in French, with 2 different methods:
- Directly when reading the file
- After creating the dataframe

```
# Directly
df3 = pd.read_csv('http://bit.ly/uforeports', header = 0, names = ['Ville', 'Couleur rapportée', 'Forme rapportée', 'Etat', 'Horodatage'])

print(df3)
```

```
# After creating
df4 = pd.read_csv('http://bit.ly/uforeports')
print(df4)

df4 = df4.rename(columns = {'City' : 'Ville', 'Colors Reported' : 'Couleur rapportée', 'Shape Reported' : 'Forme rapportée', 'State' : 'Etat', 'Time' : 'Horodatage'})
print(df4)
```
