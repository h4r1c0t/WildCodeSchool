```
def decode(string):   # Permet de replacer les caractères spéciaux..
  decodeur = {"À":"\\xc0", "Á":"\\xc1", "Â":"\\xc2", "Ã":"\\xc3", "Ä":"\\xc4", 
              "Å":"\\xc5", "Æ":"\\xc6", "Ç":"\\xc7", "È":"\\xc8", "É":"\\xc9",
              "Ê":"\\xca", "Ë":"\\xcb", "Ì":"\\xcc", "Í":"\\xcd", "Î":"\\xce", 
              "Ï":"\\xcf", "Ð":"\\xd0", "Ñ":"\\xd1", "Ò":"\\xd2", "Ó":"\\xd3", 
              "Ô":"\\xd4", "Õ":"\\xd5", "Ö":"\\xd6", "×":"\\xd7", "Ø":"\\xd8", 
              "Ù":"\\xd9", "Ú":"\\xda", "Û":"\\xdb", "Ü":"\\xdc", "Ý":"\\xdd", 
              "Þ":"\\xde", "ß":"\\xdf", "à":"\\xe0", "á":"\\xe1", "â":"\\xe2", 
              "ã":"\\xe3", "ä":"\\xe4", "å":"\\xe5", "æ":"\\xe6", "ç":"\\xe7", 
              "è":"\\xe8", "é":"\\xe9", "ê":"\\xea", "ë":"\\xeb", "ì":"\\xec",
              "í":"\\xed", "î":"\\xee", "ï":"\\xef", "ð":"\\xf0", "ñ":"\\xf1", 
              "ò":"\\xf2", "ó":"\\xf3", "ô":"\\xf4", "õ":"\\xf5", "ö":"\\xf6", 
              "÷":"\\xf7", "ø":"\\xf8", "ù":"\\xf9", "ú":"\\xfa", "û":"\\xfb", 
              "ü":"\\xfc", "ý":"\\xfd", "þ":"\\xfe", "ÿ":"\\xff"}
  for key in decodeur.keys():
    string = string.replace(decodeur[key], key)
  return string
```

```
actors_list = []
for i in range(len(actors_lists)):
  for actor in actors_lists[i][:]:
    actors_list.append(decode(actor))
```

## *Counting actors*

```
import statistics

statistics.mode(actors_list)  # nous donne un seul nom
```

```
def mode(list):
  counter = {x: list.count(x) for x in set(list)}
  mode = [0]
  max_count = 0 
  for key, value in counter.items():
    if value > max_count: 
      mode[0] = key
      max_count = value
    elif value == max_count:
      mode.append(key)
  return counter, mode

counts, mode = mode(actors_list)
mode  # nous donne une liste de 3 noms à égalité.
```
