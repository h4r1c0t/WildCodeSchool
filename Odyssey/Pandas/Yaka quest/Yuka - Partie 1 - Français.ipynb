# Mission 3

Crée une fonction qui prend en paramètre un dataframe, et affiche le taux de remplissage de chacune de ces colonnes sous forme de pourcentage. C'est à dire :
- colonne 1 : 73%
- colonne 2 : 99%

```python
 # méthode #1
 def filling(data):
  dic = {}
  for i in range(len(data.columns)):
    data_col = data.iloc[:, i]
    dic[i+1] = round(sum(data_col.notna()) / len(data_col) * 100, 2)
  return dic

fill_col = filling(df)
print(filling(df))
```

```python
# méthode #2
def filling2(data):
  return  data.count()  / len(data) *100

fill_col2 = filling2(df)  
print(fill_col2)
```

# Mission 4

Quel est le nombre de colonne appartenant à chaque quartile des taux de remplissage ? Autrement dit, combien de colonne sont remplies à 0%, entre 0 et 25%, entre 25 et 50%...

```python
print('     fill  =   0% ->',
      len([key for key, value in fill_col.items() if value == 0]))

print(' 0 < fill <=  25% ->',
      len([key for key, value in fill_col.items() if 0 < value <= 25]))

print('25 < fill <=  50% ->',
      len([key for key, value in fill_col.items() if 25 < value <= 50]))

print('50 < fill <=  75% ->',
      len([key for key, value in fill_col.items() if 50 < value <= 75]))

print('75 < fill <= 100% ->',
      len([key for key, value in fill_col.items() if 75 < value <= 100]))
```

# Mission 6

Un dataset trop pauvre en data n'est pas vraiment intéressant. Après avoir analyser les colonnes, nous nous sommes rendu compte que nous pouvons supprimer les colonnes avec moins de 60% de valeurs non nulles, sauf les colonnes pnns_groups_1 et pnns_groups_2, nutrition-score-fr_100g et nutrition-score-uk_100g. En effet ces 2 colonnes vont nous être utiles par la suite.

Si ce n'est pas déjà fait, supprime également les colonnes suivantes car elles ne sont pas vraiment utiles pour répondre à notre problématique :
* creator
* created_t
* created_datetime
* last_modified_t
* brands_tags
* countries
* countries_tags
* additives_tags
* states
* states_tags

```python
# Taux de remplissage des colonnes 
fill_fr_col = filling(frenchDF) 

# Colonnes à garder 
keepCols = ['pnns_groups_1', 'pnns_groups_2', 'nutrition-score-fr_100g', 'nutrition-score-uk_100g']

eraseCols = ['creator', 'created_t', 'created_datetime', 'last_modified_t', 
                 'brands_tags', 'countries', 'countries_tags', 'additives_tags', 
                 'states', 'states_tags'] # Liste des noms des colonnes qu'on souhaite supprimer

keepList1 = [df.columns.get_loc(col) for col in keepCols if col in frenchDF] # On récupère leurs index

eraseList1 = [df.columns.get_loc(col) for col in eraseCols if col in frenchDF]
# New DF avec les colonnes que l'on garde
greatDF = df.iloc[:, [key for key, value in fill_fr_col.items() if key not in eraseList1 and value >= 60 or key in keepList1]]
```
